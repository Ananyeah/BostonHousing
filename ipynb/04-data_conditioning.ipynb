```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st

from IPython.display import display
%matplotlib inline
```

```python

#Acquire the Data
hous_df = pd.read_csv('../data/housing.csv', names = [ "CRIM","ZN","INDUS", "CHAS","NOX", "RM", "AGE", "DIS","RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"], delimiter=r"\s+")
hous_df.shape
```

```python
hous_samp_df = hous_df.sample(5)
hous_df.describe().T
```

```python
samp_stats = hous_samp_df.describe().T
samp_stats['skew'] = st.skew(hous_samp_df)
samp_stats['kurt'] = st.kurtosis(hous_samp_df)
samp_stats
```

```python
stats = hous_df.describe().T
stats['skew'] = st.skew(hous_df)
stats['kurt'] = st.kurtosis(hous_df)
stats
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
hous_sc = scaler.fit_transform(hous_df)
hous_sc_df = pd.DataFrame(hous_sc, columns=hous_df.columns)

sc_stats = hous_sc_df.describe().T
sc_stats['skew'] = st.skew(hous_sc_df)
sc_stats['kurt'] = st.kurtosis(hous_sc_df)
display(stats)
display(sc_stats)
```

```python
fig = plt.figure(figsize=(10,10))
for i, col in enumerate(hous_df.columns):
    fig.add_subplot(7,2,1+i)
    sns.distplot(hous_df[col], label=col)
    plt.axvline(hous_df[col].mean(), c='red')
    plt.axvline(hous_df[col].median(), c='black')
    plt.legend()
```

```python
fig = plt.figure(figsize=(10,10))
for i, col in enumerate(hous_sc_df.columns):
    fig.add_subplot(7,2,1+i)
    sns.distplot(hous_sc_df[col], label=col)
    plt.axvline(hous_sc_df[col].mean(), c='red')
    plt.axvline(hous_sc_df[col].median(), c='black')
    plt.legend()
```

```python
import scipy.stats as st
```

```python
for col in hous_sc_df.columns:
    original_col_skewtest = st.skewtest(hous_df[col])
    scaled_col_skewtest = st.skewtest(hous_sc_df[col])
    print("{}\norig skew test: {} \nscaled skew test: {}\n\n".format(col, 
                                                                     original_col_skewtest,
                                                                     scaled_col_skewtest))
```

```python
# Removing NAs
hous2_df = hous_df.copy()
hous2_df= hous2_df.dropna(axis=0, how='any')
hous2_df
```

```python
#Removing 0's since log transform does not work with 0?
#Also the one's with 0s are categorical
hous2_df = hous_df.copy().drop(['MEDV', 'CHAS', 'ZN'], axis=1)
```

```python
hous_log_df = np.log(hous2_df)
hous_log_df
```

```python
scaler.fit(hous_log_df)
hous_log_sc = scaler.transform(hous_log_df)
hous_log_sc_df = pd.DataFrame(hous_log_sc, columns=hous2_df.columns)
```

```python
for col in hous_log_df.columns:
    original_col_skewtest = st.skewtest(hous_df[col])
    scaled_col_skewtest = st.skewtest(hous_sc_df[col])
    original_log_col_skewtest = st.skewtest(hous_log_df[col])
    scaled_log_col_skewtest = st.skewtest(hous_log_sc_df[col])
    print("""{}
    orig:       {} 
    scaled:     {}
    orig log:   {}
    scaled log: {}
    
    """.format(col, 
               original_col_skewtest,
               scaled_col_skewtest, 
               original_log_col_skewtest,
               scaled_log_col_skewtest))
```

```python
fig = plt.figure(figsize=(10,10))
for i, col in enumerate(hous_sc_df.columns):
    fig.add_subplot(7,2,1+i)
    sns.distplot(hous_sc_df[col], label=col)
    plt.axvline(hous_sc_df[col].mean(), c='red')
    plt.axvline(hous_sc_df[col].median(), c='black')
    plt.legend()
```

```python
fig = plt.figure(figsize=(10,10))
for i, col in enumerate(hous_log_sc_df.columns):
    fig.add_subplot(7,2,1+i)
    sns.distplot(hous_log_sc_df[col], label=col)
    plt.axvline(hous_log_sc_df[col].mean(), c='red')
    plt.axvline(hous_log_sc_df[col].median(), c='black')
    plt.legend()
```
