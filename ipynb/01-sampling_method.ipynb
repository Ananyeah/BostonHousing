```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st

from IPython.display import display
%matplotlib inline
```

```python
#Acquire the Data

hous_df = pd.read_csv("data/housing.csv", names = [ "CRIM","ZN","INDUS", "CHAS","NOX", "RM", "AGE", "DIS",
                                                     "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"
                                                    ], delimiter=r"\s+")
hous_df.shape
```

```python
stats = hous_df.describe().T
stats['skew'] = st.skew(hous_df)
stats['kurt'] = st.kurtosis(hous_df)
stats
```

It looks like the CRIM and CHAS is most skewed, and have a positive skew which means that there could be outliers but most of the data is on the negative side. 

Kurtosis is to skew what variance is to mean?
It's a measure of shape

```python
np.random.seed(42)
sample_1pct_1 = hous_df.sample(5)
sample_1pct_1
```

```python
sample_1pct_1 = hous_df.sample(5)
```

```python
(sample_1pct_1.mean() - stats['mean'])/stats['std']
```

```python
sample_1pct_10 = hous_df.sample(50)
```

```python
(sample_1pct_10.mean() - stats['mean'])/stats['std']
```

```python
sample_means = []
for _ in range(50):
    sample_means.append(hous_df.sample(50).mean())

sample_means = np.array(sample_means)
(sample_means.mean(axis=0)-stats['mean'])/stats['std']
```

Looks like 10% of the data , with means of samples repeated 50 times is closer to the population
