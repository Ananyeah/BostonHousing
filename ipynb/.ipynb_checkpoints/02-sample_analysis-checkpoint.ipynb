```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st

from IPython.display import display
%matplotlib inline
```

```python
#Acquire the Data

hous_df = pd.read_csv('../data/housing.csv', names = [ "CRIM","ZN","INDUS", "CHAS","NOX", "RM", "AGE", "DIS","RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"], delimiter=r"\s+")
hous_df.shape
```

```python
stats = hous_df.describe().T
stats['skew'] = st.skew(hous_df)
stats['kurt'] = st.kurtosis(hous_df)
stats
```

```python
np.random.seed(42)
sample_1pct_1 = hous_df.sample(5)
sample_1pct_1
```

```python
#Z-scores
cols = list(sample_1pct_1.columns)
sample_1pct_1[cols]
```

```python
for col in cols:
    col_zscore = col + '_zscore'
    sample_1pct_1[col_zscore] = (sample_1pct_1[col] - sample_1pct_1[col].mean())/sample_1pct_1[col].std(ddof=0)
sample_1pct_1
```

```python
sample_1pct_1_z = sample_1pct_1.iloc[:,-14:]
sample_1pct_1_z
```

```python
ax0 = sns.heatmap(sample_1pct_1.iloc[:,:14])
```

That's a terrible heatmap. The higher values in some columns fade out the other columns which could be equally significant.

```python
ax = sns.heatmap(sample_1pct_1_z)
```

Z-score is a better way to compare columns.

#Speculation

452 has a high criminal score (in dark brown) and a medv score that light. That's interesting
452 compared to other crim score is dark while the rest is a light color. 452 might be an outlier.

```python
pd2= pd.melt(
    sample_1pct_1.reset_index(),
    var_name='cols',
    value_vars=list(sample_1pct_1_z.columns),
    value_name='z-score'
)

pd2
```

```python
#Barplot of your sample points and of the sample point z-scores.
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
for i, col in enumerate(sample_1pct_1.columns):
    fig.add_subplot(7,2,1+i)
    sns.distplot(sample_1pct_1[col], label=col)
    plt.legend()
```
