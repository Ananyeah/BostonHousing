```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st

from IPython.display import display
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
#Acquire the Data

hous_df = pd.read_csv('../data/housing.csv', names = [ "CRIM","ZN","INDUS", "CHAS","NOX", "RM", "AGE", "DIS","RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"], delimiter=r"\s+")
hous_df1= hous_df.copy()
#hous_df1.drop('MEDV')
hous_df1.drop(['MEDV'], axis=1)
```

## Dropped the target variable here because a correlation to the target variable is actually good and does not indicate redundancy.

```python
# R2 is a measure of how much of the variance in y is explained by the model, f.

def calculate_r_2_for_feature(data,feature):
    new_data = data.drop(feature, axis=1)

    X_train, \
    X_test,  \
    y_train, \
    y_test = train_test_split(
        new_data,data[feature],test_size=0.25
    )

    regressor = KNeighborsRegressor()
    regressor.fit(X_train,y_train)

    score = regressor.score(X_test,y_test)
    return score
```

```python
calculate_r_2_for_feature(hous_df1,'CRIM')
```

```python
calculate_r_2_for_feature(hous_df1,'CRIM')
```

```python
def mean_r2_for_feature(data, feature):
    scores = []
    for _ in range(100):
        scores.append(calculate_r_2_for_feature(data, feature))
        
    scores = np.array(scores)
    return scores.mean()
```

```python
corr = hous_df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, 0)] = True
plt.subplots(figsize=(20,15))
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, square=True, annot=True,
                     cmap='RdBu', fmt='+.3f')
    plt.xticks(rotation=45, ha='center')
```
